```python
# 텍스트 정제 함수
def clean_Description(Description):
    Description = Description.replace(".", "").strip()
    Description = Description.replace("·", " ").strip()
    pattern = '[^ ㄱ-ㅣ가-힣|0-9]+'
    Description = re.sub(pattern=pattern, repl='', string=Description)
    return Description

# 문장 토큰화 함수
def get_nouns(tokenizer, sentence):
    parsed = tokenizer.parse(sentence)
    words = [line.split("\t") for line in parsed.split("\n") if line and line != "EOS"]
    nouns = [s for s, t in words if t.startswith(('NNG', 'NNP', 'VA', 'XR')) and len(s) > 1]
    return nouns
# 데이터프레임 토큰화
def tokenize(df):
    tokenizer = MeCab.Tagger("-r /opt/homebrew/etc/mecabrc")  # MeCab 객체 한 번만 생성
    processed_data = []
    for sent in tqdm(df['Description']):
        sentence = clean_Description(str(sent).replace("\n", "").strip())  # 전처리 함수 호출
        processed_data.append(get_nouns(tokenizer, sentence))
    return processed_data
```
