```python
from newspaper import Article
import requests
import pandas as pd
from datetime import datetime, timedelta

def check_article_exists(url):
    """기사 존재 여부를 HEAD 요청으로 체크"""
    res = requests.head(url, headers={"User-Agent":"Mozilla/5.0"})
    return res.status_code == 200

def crawl_articles_by_date(date_str, start_num=48000, end_num=49000):
    """
    특정 날짜 기사 크롤링 (대략적인 번호 범위 필요)
    date_str: YYYYMMDD 문자열
    """
    results = []
    base_url = "https://www.hankyung.com/article/"
    
    for num in range(start_num, end_num):
        link = f"{base_url}{date_str}{num}"
        if not check_article_exists(link):
            continue
        
        try:
            article = Article(link, language='ko')
            article.download()
            article.parse()
            results.append({
                "date": date_str,
                "url": link,
                "title": article.title,
                "text": article.text,
                "publish_date": article.publish_date
            })
            print("수집:", link)
        except Exception as e:
            print("에러:", link, e)
            
    return results

# 예시: 2025년 7월 31일 기사 크롤링
data = crawl_articles_by_date("20250731", start_num=48000, end_num=48350)
df = pd.DataFrame(data)
df.head(50)
```
