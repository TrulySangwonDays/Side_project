```python
import pandas as pd
import numpy as np

# 정규식 활용 -> 특수문자 제거
import re

# 한글 형태소 분석기
from konlpy.tag import Okt
```

```python
raw_data = pd.read_csv("/Users/leesangwon/Documents/ThemeStock_file/hankyung_i_20250808.csv")   
raw_data.head()
```

```python
# 완전일치 중복기사 제거
df = raw_data.drop_duplicates(subset=['title', 'text'], keep='first').copy()

# title에 [속보] 또는 [포토] 포함 시 text를 NaN으로 변경
df.loc[df['title'].str.contains(r'\[속보\]|\[포토\]', regex=True), 'text'] = np.nan
```

```python
df
```

```python
titles = df['title'].fillna('')
texts = df['text'].fillna('')

titles_cleaned = [
    re.sub(r'[^가-힣A-Za-z0-9%&\$₩.]', ' ', t) for t in titles
]
texts_cleaned = [
    re.sub(r'[^가-힣A-Za-z0-9%&\$₩.]', ' ', t) for t in texts
]
```

```python
texts_cleaned
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# 1) TF-IDF 변환
tfidfv = TfidfVectorizer()
testtest = tfidfv.fit_transform(titles_cleaned).toarray()

# 2) DataFrame 생성
df_count = pd.DataFrame(testtest, columns=tfidfv.get_feature_names_out())

df_count
```

```python
# NaN 제거 후 하나의 문장으로 합치기
title_all = " ".join(df['title'].dropna().tolist())
text_all = " ".join(df['text'].dropna().tolist())
```

```python
# 한글, 영문, 숫자만 포함 (특수문자 삭제. 단, &/$/%/₩는 살리기)
text_all = re.sub(r'[^가-힣A-Za-z0-9%&\$₩.]', ' ', text_all)
```

```python
text_all
```

```python
# 형태소 분석기 생성
okt = Okt()

# 토큰화
tokens = okt.morphs(text_all)
tokens

# 불용어 불러오기
with open("/Users/leesangwon/Documents/ThemeStock_file/stopwords_kor.txt", "r", encoding="utf-8") as f:
    stopwords = f.read()

# 불용어 제거
tokens = [word for word in tokens if not word in stopwords]
tokens = " ".join(tokens)

# 다시 토큰화
cleaned_tokens = okt.morphs(tokens)
cleaned_tokens
```

```python
# 품사 태깅
pos = okt.pos(tokens)
# 출력
pos
```
